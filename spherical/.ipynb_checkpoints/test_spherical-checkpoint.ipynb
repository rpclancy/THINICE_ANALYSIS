## Test if anomalies add up to 0

```python
# Set variable
nc_file = '/home/disk/sipn/rclancy/ecmwf/pf/predictability/Z500/Z500_RMSE.nc'
var_name = 'Z500_RMSE'
var_date_name = 'Z500_RMSE_dates'
t_ind = 4

# Get variables
var, var_dates =reu.read_nc_var(nc_file, var_name, var_date_name)

# Get associated dates
var_year, var_month, var_day = reu.get_y_mo_d(var_dates)

# Get lat/lon
latitude_ecmwf, longitude_ecmwf, latmesh_ecmwf, lonmesh_ecmwf = reu.get_ecmwf_lat_lon(nc_file)

# Possibly select timestep
if var_name[-1] == 'E': #basically RMSE ends in an E
    var = var[t_ind,:,:,:]

# Calculate variable anomalies
var_anom, var_anom_scaled = reu.calculate_clim_anoms(var, var_dates)
print(var_anom.shape)

fig=plt.figure(figsize=(8, 4), dpi= 80, facecolor='w', edgecolor='k')
ax = plt.axes(projection=ccrs.Robinson())
ax.set_extent([-180, 180, 60, 90], crs=ccrs.PlateCarree())
ax.set_aspect('auto')
ax.coastlines(linewidth=2)
plt.pcolormesh(lonmesh_ecmwf, latmesh_ecmwf, np.nanmean(var_anom,2), transform = ccrs.PlateCarree(),cmap='jet')
plt.colorbar(fraction=0.023, pad=0.04, label = 'var')
A, B = plt.gci().get_clim()
cmax=np.abs(np.array([A, B])).max()
plt.clim(cmax/2, cmax)
plt.show()
```

### Test generate_sic_monthly

```python
# Set variable
nc_file = '/home/disk/sipn/rclancy/ecmwf/pf/predictability/SIC/SIC.nc'
var_name = 'SIC'
var_date_name = 'SIC_dates'

# Get variables
var, var_dates =reu.read_nc_var(nc_file, var_name, var_date_name)
# Get associated dates
var_year, var_month, var_day = reu.get_y_mo_d(var_dates)

SIC_present=np.zeros([np.unique(var_year).size, 12, 20, 240])
for i,y in enumerate(np.unique(var_year)):
    for m in range(1,13):
        mo_ind = np.array(var_year==y) & np.array(var_month==m)
        SIC_present[i,m-1,:,:] = np.nanmean(var[:,:,mo_ind],2)

SIC_present[SIC_present>0]=1

fig=plt.figure(figsize=(8, 4), dpi= 80, facecolor='w', edgecolor='k')
ax = plt.axes(projection=ccrs.Robinson())
ax.set_extent([-180, 180, 60, 90], crs=ccrs.PlateCarree())
ax.set_aspect('auto')
ax.coastlines(linewidth=2)
plt.pcolormesh(lonmesh_ecmwf, latmesh_ecmwf, SIC_present[10,8,:,:], transform = ccrs.PlateCarree(),cmap='cool')
plt.colorbar(fraction=0.023, pad=0.04, label = 'var')
A, B = plt.gci().get_clim()
cmax=np.abs(np.array([A, B])).max()
plt.clim(0, cmax)
plt.show()

SIC_present.shape
np.nanmean(SIC_present,0).shape
```
